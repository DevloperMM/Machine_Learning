```python
b = [(np.random.randint(1, 23)) for i in range(15)]
b
```

```python
print(max(b))
print(min(b))
```

```python
xt = np.arange(22)
xt
```

```python
plt.hist(b, edgecolor = 'black')
plt.xticks(xt)
plt.show()
```

```python
plt.hist(b, bins = 20, edgecolor = 'black')
plt.xticks(xt)
plt.show()

# bin include the interval as [a, b) and last bin as [a, b]
```

```python
plt.hist(b, bins = [1, 4, 8, 12, 15, 20], edgecolor = 'black')
plt.axis([0, 25, 0, 10])
plt.show()

# here we have 5 bins as [1, 3) and so on and last is [15, 20]
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density = 1, facecolor='green', alpha=0.75)

# add a 'best fit' line
# y = mlab.normpdf( bins, mu, sigma)
# l = plt.plot(bins, y, 'r--', linewidth=1)

# investigate about attribute normed in plt.hist

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

# to save the graph image in the same directory
# plt.savefig("figure.png")
plt.show()
```
