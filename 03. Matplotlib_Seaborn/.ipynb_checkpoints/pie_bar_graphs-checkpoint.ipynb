## Pie Graphs

```python
import matplotlib.pyplot as plt
```

```python
sizes = [3, 4, 5, 7]
labels = ['A', 'B', 'C', 'D']
colors = ['pink', 'yellow', 'orange', 'green']
explode = [0.1, 0, 0.3, 0]

plt.pie(sizes, colors = colors, labels = labels, autopct = '%.2f%%', explode = explode)

plt.title("Split among classes")
plt.axis('equal') # by default also
plt.show()
```

```python
sizes = [3, 4, 5, 7]
labels = ['A', 'B', 'C', 'D']
colors = ['pink', 'yellow', 'orange', 'green']

plt.pie(sizes, colors = colors, labels = labels, autopct = '%.2f%%', counterclock = False, startangle = 45)

plt.title("Split among classes")
plt.axis('equal') # by default also
plt.show()
```

## Bar Graphs

```python
year = ['year - 2012','year - 2013','year - 2014','year - 2015','year - 2016','year - 2017']
salary = [12, 13, 14, 17, 19, 20]
population = [100, 160, 220, 280, 340, 400]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(year, population) # probably the default width is 0.8
plt.show()
```

```python
plt.bar(year, population, width = 0.4)
plt.xticks(rotation = 15)
plt.yticks(rotation = 15)
plt.show()
```
