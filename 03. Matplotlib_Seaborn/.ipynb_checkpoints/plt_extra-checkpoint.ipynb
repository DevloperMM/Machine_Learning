```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.rand(5)

plt.figure(figsize = (7, 4))
plt.bar(x,y)

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
plt.show()
```

```python
plt.barh(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
plt.show()
```

```python
x = np.linspace(0, 2*(math.pi), 25)
x
```

```python
y = np.sin(x)
y
```

```python
plt.figure(figsize=(5, 5))
plt.plot(x,y,'--b')
plt.xlabel("this is x label")
plt.ylabel("this is y label")
plt.title("this is title ")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)

plt.show()
```
