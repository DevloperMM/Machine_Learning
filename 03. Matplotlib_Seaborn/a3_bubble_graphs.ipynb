```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year = [2012, 2013, 2014, 2015, 2016, 2017]
salary = [12, 13, 14, 17, 19, 20]
population = np.linspace(100, 400, 6, dtype = int)
```

```python
population
```

```python
plt.scatter(year, salary, s = population, c = 'r')
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year, salary, s = population, c = 'r', alpha = 0.2)
plt.title("Bubble Chart")
plt.show()
```

```python
c = np.arange(len(year))
```

```python
plt.scatter(year, salary, s = population, c = c, alpha = 0.2, edgecolor = 'black', marker = "^")
plt.title("Bubble Chart")
plt.show()
```

```python
plt.scatter(year, salary, s = population, c = c, alpha = 0.2, edgecolor = 'black', marker = "^")
plt.title("Bubble Chart")
plt.text(year[0]+0.1, salary[0]+0.1, population[0])
plt.text(year[4]+0.1, salary[4]+0.1, population[4])
plt.show()
```
