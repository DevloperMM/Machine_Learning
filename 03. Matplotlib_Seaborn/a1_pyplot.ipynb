```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [3, 6, 9, 12]

plt.scatter(x, y)
plt.show() # to show the graph
```

```python
x = [1, 2, 3, 4]
y = [3, 6, 9, 12]

x2 = [3, 4, 5]
y2 = [2, 4, 6]

plt.plot(x2, y2, "r-.")
plt.plot(x, y)
plt.show()
```

```python
import numpy as np
# x = np.array([1, 3, 4, 5])
x = np.arange(0, 5, 0.1)
y = x ** 3
plt.plot(x, y, "bo")
plt.show()
```

```python
a = [2, 3, 4]
plt.plot(a)
plt.show()

# values of a goes to y and x take 0, 1, 2 and so on by default
```

```python
import numpy as np
# x = np.array([1, 3, 4, 5])
x = np.arange(0, 5, 0.1)
y = x ** 3
y2 = x ** 2

# plt.plot(x, y, color = 'black', marker = 'o')

plt.plot(x, y, color = 'pink', linewidth = 5, label = "X^3")
plt.plot(x, y2, color = 'orange', linewidth = 5, label = "X^2")

#plt.axis([0, 10, 0, 200])

plt.grid()

plt.text(2, 80, "text", fontsize = 15)

plt.xlabel("X")
plt.ylabel("Y")

plt.legend()
plt.title("Matplotlib demo")
plt.show()
```

```python
a = [2, 3, 4]
b = [4, 9, 16]
plt.plot(a, 'p-', antialiased = True)
plt.plot(b, 'p-', antialiased = False)
plt.show()
```
