```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# create a dataframe with a categorical feature

df = pd.DataFrame({
    'color' : ['red', 'blue', 'green', 'green', 'blue', 'red']
})
```

```python
df
```

```python
# create instance of one hot encoder

ohe = OneHotEncoder()
```

```python
ohe.fit_transform(df[['color']]).toarray()

# probably sorted alphabetwise b, g, r
```

```python
ohed = ohe.fit_transform(df[['color']]) # ohed - one hot encoded
```

```python
import pandas as pd
ohed_df = pd.DataFrame(ohed.toarray(), columns = ohe.get_feature_names_out())
```

```python
ohe.get_feature_names_out()
```

```python
ohed_df
```

```python
pd.concat([df, ohed_df], axis = 1)
```
