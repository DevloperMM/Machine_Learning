```python
df.columns = ['sl', 'sw', 'pl', 'pw', 'flower_type']
```

###### Drop Function takes the label to delete and not the index

```python
# Method-1
df = df.drop(0)
print(df.head())

# Method-2
df.drop(3, inplace = True)
print(df.head())
```

```python
df.index # gives the label
```

```python
# If we want to remove row by index

df.drop(df.index[0], inplace = True)
df.head()
```
