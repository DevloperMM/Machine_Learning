```python
import numpy as np
```

```python
data = np.loadtxt('collected_data.csv', delimiter = ',')
```

```python
x = data[:, 0].reshape(-1, 1)
y = data[:, 1]
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(x, y)
```

```python
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
```

```python
alg1.fit(x_train, y_train)
```

```python
alg1.coef_  # to get m in y = mx + c
```

```python
alg1.intercept_  # to get c in y = mx + c
```

```python
import matplotlib.pyplot as plt
```

```python
m = alg1.coef_[0]
c = alg1.intercept_

x_line = np.arange(20, 80, 2)
y_line = m * x_line + c

plt.scatter(x_train, y_train)
plt.plot(x_line, y_line, 'r-.')
plt.axis([20, 80, 20, 120])
plt.show()
```

```python
m = alg1.coef_[0]
c = alg1.intercept_

x_line = np.arange(20, 80, 2)
y_line = m * x_line + c

plt.scatter(x_test, y_test)
plt.plot(x_line, y_line, 'r-.')
plt.axis([20, 80, 20, 120])
plt.show()
```

```python
s1 = alg1.score(x_test, y_test)
s2 = alg1.score(x_train, y_train)
s1, s2
```
